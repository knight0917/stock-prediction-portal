```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Types of Plots

## Line Plot

```python
x = [1,2,3,4,5]
y = [5,28,35,45,50]
```

```python
plt.plot(x,y);
plt.title('Simple Line Plot');
plt.xlabel('x-axis');
plt.ylabel('y-axis');
plt.grid(True)
```

## Bar Chart

```python
months = ['jan', 'feb', 'mar', 'Apr', 'may']
revenue = [100,140,130,103,93]
```

```python
plt.figure(figsize=(10,3));
plt.bar(months,revenue);
plt.title('Monthly Revenue Bar chart');
plt.xlabel('--Months--');
plt.ylabel('--Revenue--');
```

## Histogram

```python
marks50 = np.random.randint(0, 100, 50)
marks50
```

```python
plt.hist(marks50);
plt.xlabel('Marks')
plt.ylabel('No. of Students')
```

```python
plt.hist(marks50, bins=5);
```

## Scatter Plot

```python
heights = [65,78,23,90,45]
weights = [150,160,30,230,170]
```

```python
plt.scatter(heights, weights);
plt.title('Heights Vs Weights');
plt.xlabel('Heights in inches');
plt.ylabel('Weights in inches');
```

## subPlots and legends

```python
days = [1,2,3,4,5]
originalPrice = [10,20,30,40,45]
predictedPrice = [8,22,34,40,44]
```

```python
plt.plot(days,originalPrice, 'b-o', label='Original Price');
plt.plot(days, predictedPrice, 'r-o', label='Predicted Price');
plt.legend()
plt.title('Plot with legends');
plt.xlabel('Weekdays');
plt.ylabel('Price');
plt.savefig('prediction.png', dpi=300)  #this save the image
```

## subPlots

```python
# first subplot- original price
plt.subplots(1)
plt.plot(days,originalPrice, 'b-o', label='Original Price');
plt.legend()
plt.title('Original price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');


# second subplot- Predicted price
plt.subplots(1)
plt.plot(days, predictedPrice, 'r-o', label='Predicted Price');
plt.legend()
plt.title('Pridicted price over time');
plt.xlabel('Weekdays');
plt.ylabel('Price');


# third subplot- diff. between original & predicted price
priceDifference = [o-p for o,p in zip(originalPrice, predictedPrice)]
plt.subplots(1)
plt.bar(days,originalPrice, color='purple', label='Price difference');
plt.legend()
plt.title('Difference between original & predicted price');
plt.xlabel('Days');
plt.ylabel('Price difference');
```
